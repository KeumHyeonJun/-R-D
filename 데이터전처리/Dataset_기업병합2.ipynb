```python
data = df_final

data.drop(['STP_Code_1','STP_Code_2','National_Strategy_1','6T_1','Start_date','End_date'], axis = 1, inplace = True)

data['Duration'] = data['Duration'].astype(str)


def remove_days(x):
    return x.replace(' days','')

data['Duration'] = data['Duration'].apply(remove_days)
data['Duration']

data['Year'] = data['Year'].astype(int)

data.drop(['Keywords','Proj_Perf_ID'],axis = 1, inplace=True)

data['R&D_Fund'] =data['R&D_Fund'].astype(float)
data['Duration'] =data['Duration'].astype(float)
df_log = data[['R&D_Fund', 'Duration']]
eps = 0.001 
df_log['Log R&D_Fund'] = np.log(df_log['R&D_Fund']+eps)
df_log['Log Duration'] = np.log(df_log['Duration']+eps)
df_log.drop(['R&D_Fund','Duration'], axis = 1 , inplace = True)
data.drop(['R&D_Fund','Duration'], axis = 1 , inplace = True)

data = pd.concat([data, df_log], axis = 1)
data.rename(columns = {'Log R&D_Fund' : 'Log_RnD_Fund'}, inplace = True)
data.rename(columns = {'Log Duration' : 'Log_Duration'}, inplace = True)

data = data.astype(str)

data['Log_RnD_Fund'] = data['Log_RnD_Fund'].astype(float)
data['Log_Duration'] = data['Log_Duration'].astype(float)
data['N_of_SCI'] = data['N_of_SCI'].astype(float)
data['N_of_Paper'] = data['N_of_Paper'].astype(float)
data['N_Patent_App'] = data['N_Patent_App'].astype(float)
data['N_Patent_Reg'] = data['N_Patent_Reg'].astype(float)
data['N_of_Korean_Patent'] = data['N_of_Korean_Patent'].astype(float)
data['N_of_Inter_Patent'] = data['N_of_Inter_Patent'].astype(float)
data['Comm_Success'] = data['Comm_Success'].astype(float)
data['Comm_Success_1'] = data['Comm_Success_1'].astype(float)
data['Comm_Success_2'] = data['Comm_Success_2'].astype(float)
data['N_of_Patent'] = data['N_of_Patent'].astype(float)
data['Application_Area_1_Weight'] = data['Application_Area_1_Weight'].astype(float)
data['Application_Area_2_Weight'] = data['Application_Area_2_Weight'].astype(float)
data['Application_Area_3_Weight'] = data['Application_Area_3_Weight'].astype(float)
data['STP_Code_1_Weight'] = data['STP_Code_1_Weight'].astype(float)
data['STP_Code_2_Weight'] = data['STP_Code_2_Weight'].astype(float)

data['N_Comm_Type_1'] = data['N_Comm_Type_1'].astype(float)
data['N_Comm_Type_2'] = data['N_Comm_Type_2'].astype(float)
data['N_Comm_Type_3'] = data['N_Comm_Type_3'].astype(float)
data['N_Comm_Type_4'] = data['N_Comm_Type_4'].astype(float)
data['N_Comm_Type_5'] = data['N_Comm_Type_5'].astype(float)
data['N_Comm_Type_6'] = data['N_Comm_Type_6'].astype(float)


data.rename(columns={'6T_2':'SixT_2'}, inplace = True)
data.rename(columns={'R&D_Org':'RnD_Org'}, inplace = True)
data.rename(columns={'R&D_Stage':'RnD_Stage'}, inplace = True)
data.rename(columns={'Multi-Year':'Multi_Year'}, inplace = True)

data['Year'] = data['Year'].astype(float)
```
