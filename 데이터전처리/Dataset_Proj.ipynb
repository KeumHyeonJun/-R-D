# Duratin Rnd Fun Log 

```python
data['Duration'] = data['Duration'].astype(str)


def remove_days(x):
    return x.replace(' days','')

data['Duration'] = data['Duration'].apply(remove_days)
data['Duration']
```

```python
data['R&D_Fund'] =data['R&D_Fund'].astype(float)
data['Duration'] =data['Duration'].astype(float)
df_log = data[['R&D_Fund', 'Duration']]
eps = 0.001 
df_log['Log R&D_Fund'] = np.log(df_log['R&D_Fund']+eps)
df_log['Log Duration'] = np.log(df_log['Duration']+eps)
df_log.drop(['R&D_Fund','Duration'], axis = 1 , inplace = True)
data.drop(['R&D_Fund','Duration'], axis = 1 , inplace = True)
```

```python
data = pd.concat([data, df_log], axis = 1)
data.rename(columns = {'Log R&D_Fund' : 'Log_RnD_Fund'}, inplace = True)
data.rename(columns = {'Log Duration' : 'Log_Duration'}, inplace = True)
```
